# Preprocessing of the Dataset

🖊 Re-scaling the amount column so that the data is much more evenly following the normal distribution curve...

```python
# The amount column is heavily imbalanced
df['Amount'].hist()
```

```python
import numpy as np
data_df =  df.copy()
# The log function rescales the data such that it follows normal distribution
data_df['Amount'] = np.log(data_df['Amount'] + 1) # add one so that there is no log below zero
data_df['Amount'].hist()
```

```python
from sklearn.preprocessing import RobustScaler
new_df = data_df.copy()
new_df['Amount'] = RobustScaler().fit_transform(new_df['Amount'].to_numpy().reshape(-1,1)) # Robust scaling helps to reduce the range of values while maintaining the shape of data
time = new_df['Time']
new_df['Time'] = (time - time.min()) / (time.max()) # to make time range between 0 and 1. This will be helpful to the models.
new_df['Amount'].hist()
```

🖊 Dividing the data to training, validation and testing sets

```python
new_df = new_df.sample(frac=1, random_state=1)
train, test, val = new_df[:240000], new_df[240000:262000], new_df[262000:]
train['Class'].value_counts(), test['Class'].value_counts(), val['Class'].value_counts()
```

🖊 Transforming the sets to numpy arrays and then dividing them up to the independent **x** (time, V1-V28, amount) and dependent **y** (class) sets for training

```python
train_np, test_np, val_np = train.to_numpy(), test.to_numpy(), val.to_numpy()
train_np.shape, test_np.shape, val_np.shape
```

```python
x_train, y_train = train_np[:, :-1], train_np[:, -1 ]
x_test, y_test = test_np[:, :-1], test_np[:, -1]
x_val, y_val = val_np[:, :-1], val_np[:, -1]
```

🖊 We shall be gauging the models according to their precision and recall reports in detecting fraud cases. Precision is the ability of the model to accurately flag transactions that are indeed fraudulent, while recall is the ability of the model to not report false negatives (not reporting a transaction as non-fraudulent when it's actually fraudulent). The best model is one that scores high in both cases.

> A model low recall will be flagging fraudulent transactions as non-fraudulent.   
> A model with low precision will be flagging non-fraudulent transactions as fraudulent.



1. Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train) # Training model
logistic_model.score(x_train, y_train) # General score from training data
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, logistic_model.predict(x_val), target_names=['Not fraud', 'Fraud'])) # Score from validation data
```

2. Sequentual model (shallow neural network)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

shallow_nn = Sequential()
shallow_nn.add(InputLayer((x_train.shape[1],)))
shallow_nn.add(Dense(2, 'relu'))
shallow_nn.add(BatchNormalization())
shallow_nn.add(Dense(1, 'sigmoid'))

checkpoint = ModelCheckpoint('shallow_nn', save_best_only=True)
shallow_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

🤔 This model begins overfitting after the 4th epoch, this will lead to lower results. We shall train another upto the 4th epoch so as to get better results.

```python
shallow_nn.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=5, callbacks=checkpoint) # Training model
```

```python
# Function that takes in a model and data. It returns the predicions from the data
def neural_net_predictions(model, x):
  return (model.predict(x).flatten() > 0.5).astype(int)
```

```python
print(classification_report(y_val, neural_net_predictions(shallow_nn, x_val), target_names=['Not fraud', 'Fraud'])) # Score from validation data
```

```python
shallow_nn_i = Sequential()
shallow_nn_i.add(InputLayer((x_train.shape[1],)))
shallow_nn_i.add(Dense(2, 'relu'))
shallow_nn_i.add(BatchNormalization())
shallow_nn_i.add(Dense(1, 'sigmoid'))

checkpoint = ModelCheckpoint('shallow_nn_i', save_best_only=True)
shallow_nn_i.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

shallow_nn_i.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=4, callbacks=checkpoint) # Training model
print(classification_report(y_val, neural_net_predictions(shallow_nn_i, x_val), target_names=['Not fraud', 'Fraud'])) # Score from validation data
```

3. Random Forest Classifier model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2, n_jobs=-1)
rf.fit(x_train, y_train) # Training model
print(classification_report(y_val, rf.predict(x_val), target_names=['Not fraud', 'Fraud'])) # Score from validation data
```

4. Gradient boosting classifier model

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)
gbc.fit(x_train, y_train) # Training model
print(classification_report(y_val, gbc.predict(x_val), target_names=['Not fraud', 'Fraud'])) # Score from validation data
```

5. Linear SVC model

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(class_weight='balanced')
svc.fit(x_train, y_train) # Training model
print(classification_report(y_val, svc.predict(x_val), target_names=['Not fraud', 'fraud'])) # Score from validation data
```
